```python
import numpy as np
import matplotlib.pyplot as plt
temparture=[39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]

temparturemonth =np.arange(1,13)

```

```python
plt.plot(temparturemonth, temparture)
temp_cel=[(f-32)*.56 for f in temparture]
plt.plot(temparturemonth,temp_cel)
```

```python
plt.plot(temparturemonth, temparture, marker ='x', color='blue', label="Farhnite")
plt.plot(temparturemonth, temp_cel, marker ='s', color='red', label="cencius")
plt.legend(loc="best")
plt.legend()
```
