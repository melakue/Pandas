## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
Male_Players=purchase_data.loc[purchase_data["Gender"]=="Male"]
Female_Players=purchase_data.loc[purchase_data["Gender"]=="Female"]
Male_no=Male_Players["SN"].nunique()
Female_No=Female_Players["SN"].nunique()
nondisclosed=purchase_data.loc[purchase_data["Gender"]== "Other / Non-Disclosed"]
nondisclosed_no=nondisclosed["SN"].nunique()
pd.options.display.float_format = '${:,.2f}'.format

DF3= pd.DataFrame({"Gender":["Male","Female","Other / Non-Disclosed"],
                  "Total Count":[Male_no,Female_No,nondisclosed_no],
                  "Percentage of Players":[(Male_no/purchase_data["SN"].nunique())*100,(Female_No/purchase_data["SN"].nunique()),(nondisclosed_no/purchase_data["SN"].nunique())*100]
                                    
                  })

DF3.set_index('Gender',inplace=True)
DF3.rename(columns={"Gender":' '})

DF3
```


## Purchasing Analysis (Gender)

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins=[0,9,14,19,24,29,34,39,500]
Group_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data["age_group"]=pd.cut(purchase_data["Age"], bins, labels=Group_labels)

DF5=purchase_data.groupby("age_group").nunique()

Age_count=DF5["SN"]/DF5["SN"].sum()


DF6=pd.DataFrame({
                 "Total Count":DF5["SN"],
                 "Percentage of Players": Age_count*100
                })
DF6
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins=[0,9,14,19,24,29,34,39,500,]
Group_labels=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
purchase_data["age_group"]=pd.cut(purchase_data["Age"], bins, labels=Group_labels)
DF=purchase_data[["SN","age_group","Price"]]
Purchase_count=DF.groupby("age_group").count()
Avrage=[I/(Purchase_count["SN"].sum())*100 for I in Purchase_count["SN"]]
pd.options.display.float_format = '${:,.2f}'.format               

Total_Purchase_Value=DF.groupby("age_group")['Price'].agg('sum')
Average_Total_Price_Per_Person=[(I/Total_Purchase_Value)*100 for I in Total_Purchase_Value]
variable=DF.groupby("age_group")["Price"].sum()/DF.groupby("age_group")["SN"].nunique()

DF7=pd.DataFrame({"Purchase Count":Purchase_count["SN"],
                 "Average Purchase Price": Avrage,
                 "Total Purchase Value": Total_Purchase_Value,
                  "Avg Total Purchase Per Person":variable
                 })


DF7

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
series=purchase_data[["SN","age_group","Price"]]
Total_Purchase_value= series.groupby(["SN"]).sum().sort_values("Price", ascending=False)

DF8=pd.DataFrame({
                 "Purchase Count": series["SN"].value_counts(),
                "Average Purchase Price": Total_Purchase_value["Price"]/series["SN"].value_counts(),
               "Total_Purchase_value":Total_Purchase_value["Price"]
                  
                  })

DF8.sort_values(["Total_Purchase_value"],ascending=False).head()

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
DF9=purchase_data[["Item ID","Item Name","Price"]]
Purchase_count["Purchase_Count"]=DF9.groupby(["Item ID","Item Name"]).count()
Purchase_count["Total_Purchase_Value"]=DF9.groupby(["Item ID","Item Name"])["Price"].sum()
Purchase_count["Item_Price"]=Purchase_count["Total_Purchase_Value"]/Purchase_count["Total_Purchase_Value"]

DF10=pd.DataFrame({"Purchase Count":Purchase_count["Purchase_Count"],
           "Item Price":Purchase_count["Total_Purchase_Value"]/Purchase_count["Purchase_Count"],
           "Total Purchase Value":Purchase_count["Total_Purchase_Value"]
          })
DF10.sort_values(["Purchase Count"],ascending=False).head()
```
